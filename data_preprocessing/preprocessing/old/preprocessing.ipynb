# DATA DIR

```python
cls = ['negative', 'hemorrhagic', 'depressed', 'protruded']
lesion = {'negative' : ['negative'], 
          'hemorrhagic': ['red_spot', 'angioectasia', 'active_bleeding'],
          'depressed': ['erosion', 'ulcer', 'stricture'],
          'protruded': ['ampulla_of_vater', 'lymphoid_follicles', 'small_bowel_tumor']}
```

```python
# path of data folder
dir_data = '/mnt/disk1/project/public/SMhospital/ChoongChoong/dataset/p2_3/'
path_folder = {}

for i in cls:
    path_folder[i] = {}
    for j in lesion[i]:
        temp = dir_data + '{0}/{1}'.format(i, j)
        path_folder[i][j] = temp
        
print(path_folder)
```

```python
list_data = {}
for i in cls:
    list_data[i] = {}
    for j in lesion[i]:
        list_data[i][j] = []
        temp = os.listdir(path_folder[i][j])
        list_data[i][j].append(temp)

print(list_data)
```

# DATA PRE-PROCESSING

```python
# for papaer writing, it needs to be revised
def pre_process(img):
    img = np.array(img, dtype = 'f4')
    img_pre = img[32:544, 32:544, :]
    for i in range(100):
        for j in range(100):
            if i + j > 99:
                pass
            else :
                img_pre[i, j, :] = 0
                img_pre[i, -j, :] = 0
    return img_pre
```

```python
dir_data_pre = '/mnt/disk1/project/public/SMhospital/ChoongChoong/dataset/pre_p2_3/'
for i in cls:
    for j in lesion[i]:
        for k in list_data[i][j][0]:
            temp = cv2.imread(dir_data + '{0}/{1}/{2}'.format(i, j, k))
            temp_pre = pre_process(temp)
            cv2.imwrite(dir_data_pre + '{0}/{1}/{2}'.format(i, j, k), temp_pre)
```
